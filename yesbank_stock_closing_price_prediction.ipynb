### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
def remove_outliers_iqr(series, lower_percentile=0.05, upper_percentile=0.95):
    lower_limit = series.quantile(lower_percentile)
    upper_limit = series.quantile(upper_percentile)
    series = np.where(series < lower_limit, lower_limit, series)
    series = np.where(series > upper_limit, upper_limit, series)
    return series


for col in data.columns[1: ]:
  data[f'{col}_iqr'] = remove_outliers_iqr(data[col])
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)


from sklearn.model_selection import KFold
k = 6
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Fit the Algorithm
mse_scores = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Train the model
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate mean squared error
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Print the mean squared error for each fold
print(f"Mean Squared Errors for each fold: {mse_scores}")

# Print the average mean squared error across all folds
print(f"Average Mean Squared Error: {np.mean(mse_scores)}")
````
